```python
class Enregistrement:

    def __init__(self, temperature, humidite, precipitations, vent, pression, indice_uv):
        self.__temperature=temperature
        self.__humidite=humidite
        self.__precipitations=precipitations
        self.__vent=vent
        self.__pression= pression
        self.__indice_uv= indice_uv
    
    def __str__(self):
        return (f"Enregistrement(,temp={self.__temperature},hum={self.__humidite}, precip={self.__precipitations},vent={self.__vent}, pression={self.__pression}, UV={self.__indice_uv}, score={self.score()})")
    
    def getTemperature(self):
        return self.__temperature
    
    def getHumidite(self):
        return self.__humidite
    
    def getPrecipitations(self):
        return self.__precipitations
    
    def getVent(self):
        return self.__vent
    
    def getPression(self):
         return self.__pression
    
    def getIndiceUV(self): 
        return self.__indice_uv
    

    def setTemperature(self,v):
        self.__temperature=v

    def setHumidite(self, v):   
        self.__humidite=v

    def setPrecipitations(self, v):
        self.__precipitations=v

    def setVent(self, v):         
        self.__vent=v

    def setPression(self, v):    
        self.__pression=v

    def setIndiceUV(self, v):    
        self.__indice_uv=v


    def score(self):
        total = 0
        t = self.__temperature
        if not (isinstance(t, float) and math.isnan(t)):
            if 18 <= t <= 28:
                total += 25
            elif (10 <= t < 18) or (28 < t <= 32):
                total += 15
            else:
                total += 5
 
        h = self.__humidite
        if not (isinstance(h, float) and math.isnan(h)):
            if 40 <= h <= 70:
                total += 15
            elif (30 <= h < 40) or (70 < h <= 80):
                total += 8
            else:
                total += 3

        p = self.__precipitations
        if not (isinstance(p, float) and math.isnan(p)):
            if 0 <= p <= 2:
                total += 15
            elif 2 < p <= 10:
                total += 8
            else:
                total += 0
 
        v = self.__vent
        if not (isinstance(v, float) and math.isnan(v)):
            if 0 <= v <= 30:
                total += 15
            elif 30 < v <= 60:
                total += 8
            else:
                total += 0
 
        pr = self.__pression
        if not (isinstance(pr, float) and math.isnan(pr)):
            if 1010 <= pr <= 1025:
                total += 15
            elif (1000 <= pr < 1010) or (1025 < pr <= 1035):
                total += 8
            else:
                total += 3
 
        uv = self.__indice_uv
        if not (isinstance(uv, float) and math.isnan(uv)):
            if 0 <= uv <= 5:
                total += 15
            elif 5 < uv <= 7:
                total += 8
            else:
                total += 2
 
        return total

```

```python

class Station:
    ANNEE=[("janvier",31),
("fevrier",29),
("mars",31),
("avril",30),
("mai",31),
("juin",30),
("juillet",31),
("aout",31),
("septembre",30),
("octobre",31),
("novembre",30),
("decembre",31)]
    def __init__(self, nom, localisation):
        self.__nom = nom
        self.__localisation = localisation
        self.__data = {} 

 
    def __str__(self):
        annees = list(self.__data.keys())
        return (f"Station(nom='{self.__nom}',localisation={self.__localisation},années={annees})")
    
 
    def getNom(self):    
        return self.__nom
    
    def getLocalisation(self):  
        return self.__localisation
    
    def getData(self):       
        return self.__data
 

    def setNom(self, v):   
         self.__nom = v

    def setLocalisation(self, v):
        self.__localisation = v

    def chargerDonnees(self):
        data_annee = {mois: {} for (mois, _) in Station.ANNEE}
 
        for mois, jours in Station.ANNEE:
            data_annee[mois] = {
                jour: Enregistrement(
                    random.choice([np.nan, random.randint(0, 40)]),  
                    random.choice([np.nan, random.randint(0, 100)]),   
                    random.choice([np.nan, random.randint(0, 20)]),    
                    random.choice([np.nan, random.randint(0, 100)]), 
                    random.choice([np.nan, random.randint(980, 1040)]),
                    random.choice([np.nan, random.randint(0,11 )])     
                )
                for jour in range(1, jours + 1)
            }
 
        self.__data = data_annee
        print(f"  Données chargées pour la station '{self.__nom}.")

    def _enregistrements_mois(self, mois):
        if mois not in self.__data:
            raise ValueError(f"Mois '{mois}' inconnu.")
        return self.__data[mois]
    
    def temperatures_uniques(self, mois):
        enreg = self._enregistrements_mois(mois)
        valeurs = set()
        for e in enreg.values():
            t = e.getTemperature()
            if not (isinstance(t, float) and math.isnan(t)):
                valeurs.add(t)
        return sorted(valeurs)
 
    def temperature_moyenne(self, mois):
        enreg = self._enregistrements_mois(mois)
        valeurs = [e.getTemperature() for e in enreg.values()
                   if not (isinstance(e.getTemperature(), float)
                           and math.isnan(e.getTemperature()))]
        return round(np.mean(valeurs), 2) if valeurs else np.nan
 
    def jours_pluvieux(self, mois):
        enreg = self._enregistrements_mois(mois)
        jours = []
        for jour, e in enreg.items():
            p = e.getPrecipitations()
            if not (isinstance(p, float) and math.isnan(p)) and p > 0:
                jours.append(jour)
        return sorted(jours)
 
    def jours_extremes(self, mois):
        enreg = self._enregistrements_mois(mois)
        jours = []
        for jour, e in enreg.items():
            t = e.getTemperature()
            if not (isinstance(t, float) and math.isnan(t)):
                if t < 5 or t > 35:
                    jours.append(jour)
        return sorted(jours)
 
    def humidite_moyenne(self, mois):
        enreg = self._enregistrements_mois(mois)
        valeurs = [e.getHumidite() for e in enreg.values()
                   if not (isinstance(e.getHumidite(), float)
                           and math.isnan(e.getHumidite()))]
        return round(np.mean(valeurs), 2) if valeurs else np.nan
 
    def precipitations_totales(self, mois):
        enreg = self._enregistrements_mois(mois)
        valeurs = [e.getPrecipitations() for e in enreg.values()
                   if not (isinstance(e.getPrecipitations(), float)
                           and math.isnan(e.getPrecipitations()))]
        return round(sum(valeurs), 2)
 
    def jours_beaux(self, mois):
        enreg = self._enregistrements_mois(mois)
        return sorted([jour for jour, e in enreg.items() if e.score() >= 80])
 
    def score_moyen(self, mois):
        enreg = self._enregistrements_mois(mois)
        scores = [e.score() for e in enreg.values()]
        return round(np.mean(scores), 2) if scores else 0
 
    def mois_beau(self):
        mois_scores = {
            mois: self.score_moyen(mois)
            for mois in self.__data
        }
        return max(mois_scores, key=mois_scores.get)
 

 
    def exporter_csv(self):
        fichier1 = f"{self.__nom}.csv"
 
        with open(fichier1, mode='w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow([
                "nom_station", "annee", "mois", "jour",
                "temperature", "humidite", "precipitations",
                "vent", "pression", "indice_uv", "score"
            ])
            for mois, _ in Station.ANNEE:
                if mois not in self.__data:
                    continue
                for jour, e in sorted(self.__data[mois].items()):
                    writer.writerow([
                        self.__nom, mois, jour,
                        e.getTemperature(),
                        e.getHumidite(),
                        e.getPrecipitations(),
                        e.getVent(),
                        e.getPression(),
                        e.getIndiceUV(),
                        e.score()
                    ])
 
        print(f"  CSV exporté : {fichier1}")
        return fichier1
    
    def _str_(self):
        lat, lon, alt = self.__localisation
        nb_jours = sum(len(self.__data.get(m, {})) for m, _ in Station.ANNEE)
        return (f"Station '{self.__nom}' | Loc=({lat}, {lon}, alt={alt}m) | "
                f"{nb_jours} enregistrements")
 
```

```python
def tester_station(station, mois_test="janvier"):
    print(f"\n{'='*60}")
    print(f"  Tests pour : {station.getNom()}")
    print(f"{'='*60}")
    print(f"  Températures uniques ({mois_test})  : "
          f"{station.temperatures_uniques(mois_test)[:5]}...")
    print(f"  Température moyenne  ({mois_test})  : "
          f"{station.temperature_moyenne(mois_test)} °C")
    print(f"  Jours pluvieux       ({mois_test})  : "
          f"{station.jours_pluvieux(mois_test)}")
    print(f"  Jours extrêmes       ({mois_test})  : "
          f"{station.jours_extremes(mois_test)}")
    print(f"  Humidité moyenne     ({mois_test})  : "
          f"{station.humidite_moyenne(mois_test)} %")
    print(f"  Précipitations tot.  ({mois_test})  : "
          f"{station.precipitations_totales(mois_test)} mm")
    print(f"  Jours beaux (≥80)    ({mois_test})  : "
          f"{station.jours_beaux(mois_test)}")
    print(f"  Score moyen          ({mois_test})  : "
          f"{station.score_moyen(mois_test)}")
    print(f"  Mois le plus beau    (année)     : "
          f"{station.mois_beau()}")
```

```python
class StationMultiAnnee:

    ANNEE = [
        ("janvier", 31), ("fevrier", 29), ("mars", 31),
        ("avril", 30), ("mai", 31), ("juin", 30),
        ("juillet", 31), ("aout", 31), ("septembre", 30),
        ("octobre", 31), ("novembre", 30), ("decembre", 31)
    ]

    def __init__(self, nom, localisation):
        self.__nom = nom
        self.__localisation = localisation
        self.__data = {}

    def getNom(self):
        return self.__nom

    def getLocalisation(self):
        return self.__localisation

    def getData(self):
        return self.__data

    def setNom(self, nom):
        self.__nom = nom

    def setLocalisation(self, loc):
        self.__localisation = loc

    def chargerDonnees(self, annee):
        data_annee = {}
        for mois, jours in StationMultiAnnee.ANNEE:
            data_annee[mois] = {
                jour: Enregistrement(
                    random.choice([np.nan, random.randint(0, 40)]),
                    random.choice([np.nan, random.randint(0, 100)]),
                    random.choice([np.nan, random.randint(0, 20)]),
                    random.choice([np.nan, random.randint(0, 100)]),
                    random.choice([np.nan, random.randint(980, 1040)]),
                    random.choice([np.nan, random.randint(0, 11)])
                )
                for jour in range(1, jours + 1)
            }
        self.__data[annee] = data_annee

    def temperatures_uniques(self, annee, mois):
        temps = set()
        for jour, enr in self.__data[annee][mois].items():
            t = enr.getTemperature()
            if not np.isnan(t):
                temps.add(t)
        return temps

    def temperature_moyenne(self, annee, mois):
        vals = [enr.getTemperature() for enr in self.__data[annee][mois].values()
                if not np.isnan(enr.getTemperature())]
        return np.mean(vals) if vals else np.nan

    def jours_pluvieux(self, annee, mois):
        return [jour for jour, enr in self.__data[annee][mois].items()
                if not np.isnan(enr.getPrecipitations()) and enr.getPrecipitations() > 0]

    def jours_extremes(self, annee, mois):
        result = []
        for jour, enr in self.__data[annee][mois].items():
            t = enr.getTemperature()
            if not np.isnan(t) and (t < 5 or t > 35):
                result.append(jour)
        return result

    def humidite_moyenne(self, annee, mois):
        vals = [enr.getHumidite() for enr in self.__data[annee][mois].values()
                if not np.isnan(enr.getHumidite())]
        return np.mean(vals) if vals else np.nan

    def precipitations_totales(self, annee, mois):
        vals = [enr.getPrecipitations() for enr in self.__data[annee][mois].values()
                if not np.isnan(enr.getPrecipitations())]
        return sum(vals)

    def jours_beaux(self, annee, mois):
        return [jour for jour, enr in self.__data[annee][mois].items()
                if enr.score() >= 80]

    def score_moyen(self, annee, mois):
        scores = [enr.score() for enr in self.__data[annee][mois].values()]
        return np.mean(scores) if scores else 0

    def mois_beau(self, annee):
        meilleur_mois = None
        meilleur_score = -1
        for mois, _ in StationMultiAnnee.ANNEE:
            sm = self.score_moyen(annee, mois)
            if sm > meilleur_score:
                meilleur_score = sm
                meilleur_mois = mois
        return meilleur_mois, meilleur_score

    def exporter_csv(self):
        nom_fichier = f"{self.__nom}_multi.csv"
        with open(nom_fichier, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(["station", "annee", "mois", "jour",
                             "temperature", "humidite", "precipitations",
                             "vent", "pression", "indice_uv", "score"])
            for annee in sorted(self.__data.keys()):
                for mois, _ in StationMultiAnnee.ANNEE:
                    for jour, enr in self.__data[annee][mois].items():
                        writer.writerow([
                            self.__nom, annee, mois, jour,
                            enr.getTemperature(), enr.getHumidite(),
                            enr.getPrecipitations(), enr.getVent(),
                            enr.getPression(), enr.getIndiceUV(),
                            enr.score()
                        ])
        print(f"Fichier '{nom_fichier}' exporté avec succès.")

    def _str_(self):
        lat, lon, alt = self.__localisation
        annees = sorted(self._data.keys()) if self._data else []
        return (f"StationMulti '{self.__nom}' | Loc=({lat}, {lon}, alt={alt}m) | "
                f"Années : {annees}")
```

```python
rabat_multi = StationMultiAnnee("Station_Rabat", (34.0209, -6.8416, 75))

for an in [2023, 2024, 2025]:
    rabat_multi.chargerDonnees(an)

print(rabat_multi)

print("\nTempérature moyenne de janvier par année :")
for an in [2023, 2024, 2025]:
    print(f"  {an} : {rabat_multi.temperature_moyenne(an, 'janvier'):.2f} °C")

print("\nMois le plus beau par année :")
for an in [2023, 2024, 2025]:
    mb, ms = rabat_multi.mois_beau(an)
    print(f"  {an} : {mb} (score = {ms:.2f})")

rabat_multi.exporter_csv()
```
